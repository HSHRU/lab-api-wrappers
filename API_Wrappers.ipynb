```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import time
```

```python
#Client_ID_spotify
secrets_file = open("Client_ID_spotify.txt","r")
string = secrets_file.read()
string.split('\n')
secrets_dict={}
for line in string.split('\n'):
    if len(line) > 0:
        secrets_dict[line.split(':')[0]]=line.split(':')[1]
```

```python
#Initialize SpotiPy with user credentials
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id=secrets_dict['CID'],
                                                           client_secret=secrets_dict['C_Secret']))
```

```python
playlist = sp.user_playlist_tracks("spotify", "4STozIaL9JW1bnhRC1lHoA")
```
